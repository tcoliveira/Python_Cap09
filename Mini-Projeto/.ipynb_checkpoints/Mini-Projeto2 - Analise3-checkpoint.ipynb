## Análise 3

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python

# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
df.describe
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
fig = sns.catplot(x = 'fuelType',y="price", hue="gearbox", data=df,kind="bar", height=7, aspect=1.5)
fig.ax.xaxis.set_label_text("Tipo de combustivel",fontdict= {'size':16})
fig.ax.yaxis.set_label_text("Preco", fontdict= {'size':16})     
fig.ax.set_title("Veículos Por Marca",fontdict= {'size':18})
plt.show()
```

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price_thiago.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python

colors = ["#00e600", "#ff8c1a","#a180cc"]
fig = sns.catplot(x = 'vehicleType',y="powerPS", hue="gearbox", data=df,kind="bar",palette=colors, height=5, aspect=1.5)
fig.ax.xaxis.set_label_text("Potencia Media",fontdict= {'size':16})
fig.ax.yaxis.set_label_text("Tipo Veiculo", fontdict= {'size':16})     
fig.ax.set_title("Potência média de um veículo por tipo de veículo",fontdict= {'size':18})
plt.show()

```
